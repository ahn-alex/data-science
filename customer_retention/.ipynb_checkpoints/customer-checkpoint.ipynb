# preprocessing

### gender, income has NaN 

```python
df['gender'].value_counts()
```

### going to assume MCAR -- drop

```python
df.iloc[np.where(df['gender'].isna())]
```

```python
df.iloc[np.where(df['income'].isna())]
```

```python
ppdf = df.dropna()
```

```python
ppdf.replace({'gender' : {'Male':0, 'Female':1}}, inplace = True)
ppdf.replace({'membership' : {'No':0, 'Yes':1}}, inplace = True)
```

```python
ppdf.info()
```

```python
ppdf[['gender', 'region', 'membership', 'churn']] = ppdf[['gender', 'region', 'membership', 'churn']].astype('category')
```

```python
ppdf.info()
```

```python
temp = pd.get_dummies(ppdf['region'])
ppdf=ppdf.join(temp)
ppdf.drop('region', axis=1, inplace=True)
```

```python
ppdf
```

```python
X = ppdf.drop(columns = ['customer_id', 'churn', 'login_groups', 'purchase_bins'])
y = ppdf['churn']

Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, random_state = 0)
```

```python
label = LabelEncoder()

ytrain = label.fit_transform(ytrain)
ytest = label.transform(ytest)
```

```python
from sklearn.metrics import accuracy_score, classification_report, f1_score
```

```python
lr = LogisticRegression(random_state=0)

lr.fit(Xtrain, ytrain)
ypred = lr.predict(Xtest)

accuracy_score(ytest, ypred)
```

```python
rfc = RandomForestClassifier(random_state=0)

rfc.fit(Xtrain, ytrain)
ypred = rfc.predict(Xtest)

accuracy_score(ytest, ypred)
```

```python
gbc = GradientBoostingClassifier(random_state=0)

gbc.fit(Xtrain, ytrain)
ypred = gbc.predict(Xtest)

accuracy_score(ytest, ypred)
```
